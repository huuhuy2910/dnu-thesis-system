```sql
/****** Object:  StoredProcedure [dbo].[sp_UpdateAllCurrentGuidingCount]    Script Date: 08/10/2025 2:40:19 CH ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- Stored procedure để cập nhật lại toàn bộ CurrentGuidingCount
CREATE PROCEDURE [dbo].[sp_UpdateAllCurrentGuidingCount]
AS
BEGIN
    SET NOCOUNT ON;

    UPDATE lp
    SET CurrentGuidingCount = (
        SELECT COUNT(*)
        FROM TopicLecturers tl
        WHERE tl.LecturerProfileID = lp.LecturerProfileID
          AND tl.IsPrimary = 1
    )
    FROM LecturerProfiles lp;
END

GO
```
